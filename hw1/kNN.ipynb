# Simple kNN realisation by Watemus.

```python
import random

num_train_els = 100
num_test_els = 100
k = 3

def gen_rand_data(len_):
    return [(float(random.randint(1, 100)), float(random.randint(1, 100))) for _ in range(len_)]

train = gen_rand_data(num_train_els)

train_x = list(map(lambda x: x[0], train))
train_y = list(map(lambda x: x[1], train))

train_clf = [random.randint(0, 2) for _ in range(num_train_els)]

print('x; y; clf')
for i in range(len(train)):
    print('{}; {}; {}'.format(train_x[i], train_y[i], train_clf[i]))
```

```python
from math import sqrt

def kNN(k, train, test, clf_train):
    clf_test = [0 for i in range(len(test))]
    for i in range(len(train)):
        map_ = [0, 0, 0]
        clf_c = [(clf_train[i], sqrt((train[t][0] - test[i][0]) ** 2 + (train[t][1] - test[i][1]) ** 2)) for t in range(len(train))]
        
        clf_c.sort(key=lambda dist: dist[1])
        for j in range(k):
            map_[clf_c[j][0]] += 1
        clf_test[i] = map_.index(max(map_))
    return clf_test
```

```python
test = gen_rand_data(num_test_els)
clf = kNN(k, train, test, train_clf)
print('x; y; clf')
for i in range(num_test_els):
    print('{}; {}; {}'.format(test[i][0], test[i][1], clf[i]))
```
